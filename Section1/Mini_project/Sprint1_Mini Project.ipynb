```python
dfNanList = df.isnull().sum()
for i in range(len(dfNanList)):
    if dfNanList[i] >= 500:
        print(dfNanList.index[i]," = ", dfNanList[i])

# 결측치가 500개 이상인 열은 총 4개이다.
# 이 중에서 job_title과 job_industry_category 백수이거나 특별히 분류가 되지 않는 업종이 있을 수 있다.
# 이에 default과 Unnamed: 13는 데이터로서의 결측치도 많고 데이터 분석에 필요성이 없어 제외한다.
```

```python
df_pro = df.drop(["default","Unnamed: 13"], axis=1, inplace=False)
df_pro
```

```python
# 데이터에 잘 못 입력된 데이터 수정
df_pro["gender"] = df_pro["gender"].apply(lambda x: "Female" if x=="F" or x=="Femal" else x)
df_pro["gender"] = df_pro["gender"].apply(lambda x: "Male" if x=="M" else x)

df_pro["gender"].value_counts()
```

```python
df_pro["name"] = df_pro.apply(lambda row: row["first_name"]+"_"+row["last_name"] if pd.notnull(row["last_name"]) else row["first_name"], axis=1)
df_pro
```

```python
df_pro.columns
newList = ['customer_id', 'name','gender',
       'past_3_years_bike_related_purchases', 'DOB', 'job_title',
       'job_industry_category', 'deceased_indicator',
       'owns_car','wealth_segment']
```

```python
df_pro = df_pro[newList]
df_pro
```

```python
# 생년월일은 데이터 분석에서 나이별 데이터 상태를 볼 수 있기에 없는 경우 해당 데이터 전체를 쓸 수 없기에 제거
df_pro2 = df_pro.dropna(subset=["DOB"])
df_pro2
```

```python
def madePie(df,i):
    plt.pie(df[i].value_counts().values,labels=df[i].value_counts().index,autopct='%1.1f%%', shadow=True, startangle=140)
    plt.axis('equal')  # 원형을 유지하기 위해 비율을 같게 설정
    plt.title(i)
    plt.show()
```
